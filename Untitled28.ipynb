## **Data Preparation**

```python
y=train_dataset["Close"]
x=train_dataset.drop(columns=[ "Date", "Close"] , axis=1)
```

```python
x_train , x_val , y_train, y_val=train_test_split(x , y, test_size=0.2 , random_state=42)
```

```python
num_pipeline = Pipeline([
('std_scaler',  StandardScaler()),
                          ])
```

```python
num_attribs = list(x_train[["Open" , "High" , 'Low', "Volume", 'month', 'days']])
cat_attribs = ["day_of_week"]
full_pipeline = ColumnTransformer([
("num", num_pipeline, num_attribs),
("cat", OneHotEncoder(), cat_attribs),
])
x_train_prepared = full_pipeline.fit_transform(x_train)
```

```python
x_val_prepared=full_pipeline.fit_transform(x_val)
```

## **PyTorch Data Preparation**

```python
X = torch.tensor(x_train_prepared, dtype=torch.float32)
Y = torch.tensor(y_train, dtype=torch.float32)
dataset = TensorDataset(X, Y)
dataloader = DataLoader(dataset)
```

```python
x_val_tensor=torch.tensor(x_val_prepared, dtype=torch.float32)
y_val_tensor= torch.tensor(y_val.values, dtype=torch.float32)
dataset_val = TensorDataset(x_val_tensor, y_val_tensor)
dataloader_val = DataLoader(dataset_val)
```

## **Building Model and Trainig**

```python
class RegressionModel(nn.Module):
    def __init__(self):
        super(RegressionModel, self).__init__()
        self.fc1 = nn.Linear(11, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 32)
        self.fc4 = nn.Linear(32, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = torch.relu(self.fc3(x))

        x = self.fc4(x)
        return x
```

```python
model = RegressionModel()
```

```python
criterion = nn.MSELoss()
optimizer =Adam(model.parameters(), lr=0.0001)
```

```python
train_rmse_values = []
val_rmse_values = []
num_epochs=100
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, targets in dataloader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs.squeeze(), targets)  # Calculate MSE
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    # Calculate RMSE for training set
    train_mse = running_loss / len(dataloader)
    train_rmse = np.sqrt(train_mse)  # Calculate RMSE from MSE
    train_rmse_values.append(train_rmse)

    # Calculate RMSE for validation set
    model.eval()
    val_running_loss = 0.0
    with torch.no_grad():
        for inputs, targets in dataloader_val:
            outputs = model(inputs)
            val_loss = criterion(outputs.squeeze(), targets)
            val_running_loss += val_loss.item()

    val_mse = val_running_loss / len(dataloader_val)
    val_rmse = np.sqrt(val_mse)  # Calculate RMSE from MSE
    val_rmse_values.append(val_rmse)

    print(f"Epoch {epoch+1}/{num_epochs}, RMSE (Training): {train_rmse}")
    print(f"Epoch {epoch+1}/{num_epochs}, RMSE (Validation): {val_rmse}")
```

## **Evaluation**

```python
# Plotting
epochs = range(1, num_epochs + 1)

plt.figure(figsize=(8, 5))
plt.plot(epochs, train_rmse_values, label='Training RMSE')
plt.plot(epochs, val_rmse_values, label='Validation RMSE')
plt.title('Training and Validation RMSE')
plt.xlabel('Epochs')
plt.ylabel('RMSE')
plt.legend()
plt.grid(True)
plt.show()
```
